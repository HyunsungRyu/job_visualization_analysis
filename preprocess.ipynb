### 이 코드는 테블로 시각화를 위한 데이터 전처리 과정입니다.



crawled_data : 채용사이트를 크롤링해서 얻은 데이터



```python
import pandas as pd 
df = pd.read_csv("./crawled_data.csv", encoding="cp949")
df.head(3)
```

```python
df_cleaned = df.dropna()

df_cleaned = df_cleaned.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df_cleaned.columns = ['Job_Title', 'Company_Name', 'Location', 'Experience', 'Main_Tasks', 'Qualifications', 'Preferences']

df_cleaned.head(3)
```

```python
data_analyst = df_cleaned[df_cleaned['Job_Title'].str.contains('analyst|Data\s*Analyst|데이터\s*분석가*', case=False, na=False)]
data_scientist = df_cleaned[df_cleaned['Job_Title'].str.contains('scientist|Data\s*Scientist|데이터\s*과학자', case=False, na=False)]
```

```python
print(len(data_analyst))
data_analyst.head(3)
```

```python
import re

def convert_experience(exp):
    # 경력 X년 이상
    match = re.match(r'경력\s*(\d+)\s*년\s*이상', exp)
    if match:
        return f"{match.group(1)}~"
    
    # 경력 X년 이하
    match = re.match(r'경력\s*(\d+)\s*년\s*이하', exp)
    if match:
        return f"0~{match.group(1)}"
    
    # 경력 X년 - Y년
    match = re.match(r'경력\s*(\d+)\s*-\s*(\d+)\s*년', exp)
    if match:
        return f"{match.group(1)}~{match.group(2)}"
    
    # 신입 - 경력 X년
    match = re.match(r'신입\s*-\s*경력\s*(\d+)\s*년', exp)
    if match:
        return f"0~{match.group(1)}"
    
    # 경력 X년
    match = re.match(r'경력\s*(\d+)\s*년', exp)
    if match:
        return f"{match.group(1)}~{match.group(1)}"
    
    # 신입
    match = re.match(r'신입', exp)
    if match:
        return "0~0"
    
    return exp

data_analyst['Experience'] = data_analyst['Experience'].apply(convert_experience)
data_analyst.head(3)
```

```python
data_analyst[['Company_Name','Location','Experience']].to_csv('data_analyst_Ex.csv', index=False, encoding='cp949')
```

```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 주요 업무에 있을 만한 한국어 불용어 리스트
korean_stop_words = [
    '및', '을', '를', '에', '의', '가', '이', '은', '는', '들', '으로', '하다', '있다', '수', '등', '및', '에서', '이다',
    '그리고', '또한', '하지만', '그러나', '때문에', '따라서', '그러므로', '즉', '예를', '들면', '예를', '들어', '같은', '경우',
    '때문', '위해', '통해', '대한', '대해', '관련', '관련된', '관련하여', '관한', '관해서', '위한', '위해서', '와', '하고',
    '합니다', 'ㆍ', '따른', '하여', '합니다', '있어요', '해요', '해', '한', '필요', '주요업무', '있도록', '있습니다', '있으신', '우대사항', '과',
    '하신', '원활히', '하거나', '으신', '있으면', '더욱', '이나', '분이면', '하게', '자격요건', '할', '년', '찾아요', '함께',
    '갖추고', '로', '주실', '생기면', '보실', '하셨거나', '있게'
    
]


def remove_stop_words(text, stop_words):
    pattern = re.compile(r'\b(' + '|'.join(stop_words) + r')\b|\b(' + '|'.join(stop_words) + r')\w*\b|\b\w*(' + '|'.join(stop_words) + r')\b')
    return pattern.sub('', text)

want = ['Main_Tasks', 'Qualifications','Preferences']

for i in want:
    data_analyst[i] = data_analyst[i].apply(lambda x: remove_stop_words(x, korean_stop_words))
    for j in range(2):
        vectorizer = TfidfVectorizer(ngram_range=(j+1, j+1))
        X = vectorizer.fit_transform(data_analyst[i])

        tfidf_scores = X.sum(axis=0).A1
        tfidf_scores_df = pd.DataFrame({'term': vectorizer.get_feature_names_out(), 'score': tfidf_scores})
        tfidf_scores_df = tfidf_scores_df.sort_values(by='score', ascending=False)

        top_keywords = tfidf_scores_df.head(50 if j == 0 else 30)
        print(top_keywords)
        top_keywords.to_csv(f'{i}_TopKeywords_{j+1}.csv', index=False, encoding='cp949')
```
